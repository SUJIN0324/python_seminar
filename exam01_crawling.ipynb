```python
import requests
from bs4 import BeautifulSoup
```

## 네이버 헤드라인 뉴스 가져오기  
https://news.naver.com

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36',
    'referer': 'https://search.naver.com/search.naver',
 }

response = requests.get('https://news.naver.com', headers=headers)
```

```python
html = response.text
```

```python
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
for article in soup.select('.hdline_article_list .hdline_article_tit'):
   print(article.text.strip())
```

## 멜론 차트 목록 가져오기  
https://www.melon.com/chart/index.htm

```python
path = 'https://www.melon.com/chart/index.htm'
response = requests.get(path, headers=headers)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
singer = soup.select('.wrap_song_info .rank02 .checkEllipsis')
song = soup.select('.wrap_song_info .rank01')
```

```python
for i, s in enumerate(song):
    print(i+1,':', s.text.strip(), '-', singer[i].text.strip())
```
